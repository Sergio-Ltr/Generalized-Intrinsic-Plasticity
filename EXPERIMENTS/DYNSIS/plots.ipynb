```python
import numpy as np
import matplotlib.pyplot as plt

# Define the dynamical system
def dynamical_system(x, r):
    return r - x**2

# Parameter range
r_values = np.linspace(-2, 2, 1000)

# Initial conditions and parameters
x0 = 0.5  # Adjusted initial condition
num_iterations = 1000
transient_iterations = 200  # Increased transient iterations

# Arrays to store results
x_asymptotic = []
r_saddle = []

# Iterate over parameter range
for r in r_values:
    x = x0
    # Transient dynamics
    for _ in range(transient_iterations):
        x = dynamical_system(x, r)
    # Collect asymptotic value
    x_steady_state = []
    for _ in range(num_iterations):
        x = dynamical_system(x, r)
        x_steady_state.append(x)
    x_asymptotic.append(np.mean(x_steady_state))  # Taking mean as the asymptotic value
    r_saddle.append(r)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(r_saddle, x_asymptotic, c='b')
plt.xlabel('Parameter (r)')
plt.ylabel('Asymptotic Value of Variable (x)')
plt.title('Saddle-Node Bifurcation Diagram')
plt.grid(True)
plt.show()
```
